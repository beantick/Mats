Buatlah function dengan parameter yang akan diisi dengan argumen berupa sebuah kalimat. 

Contoh:
- Argumen: 
        teh cukup murah
- Output: 
        teh pukuc harum

Constraints: 
- Hanya kata yang memiliki panjang lebih dari 3 huruf yang dibalik.
- Input hanya berupa huruf dan spasi, tanpa tanda baca.

```python
# Insert your code here

def reverse_sentence(kalimat):
  list_kalimat = kalimat.split()

  for i in list_kalimat:
    if len(i) > 3:
      print(i[::-1], end=' ')
    else:
      print(i, end=' ')

reverse_sentence('ini adalah materi tentang function')
```

**3.**

Buatlah function untuk memutar list di dalam list yang berukuran 3x3 satu kali searah jarum jam.

Misal:

- Input
      [[1,2,3],   
      [4,5,6],
      [7,8,9]]

- Output
      [[7,4,1],   
      [8,5,2],    
      [9,6,3]]   

```python
# Insert your code here

list_angka = [[1,2,3], [4,5,6], [7,8,9]]

def rotasi(x):
  list_hasil = []
  i = len(x) - 1 # 3 - 1 = 2

  while i >= 0: # 2 1 0
    for j in range(0, len(x)):  # 0 1 2
      if (len(list_hasil) < (j + 1)):
        list_hasil.append([x[i][j]])
      else:
        list_hasil[j].append(x[i][j]) 
    
    i -= 1
  return list_hasil

print(rotasi(list_angka))

# list_angka_hasil = [[7, 4, 1], [8, 5, 2], [9, 6, 3]]
```

**4.**

Jika parameter sebuah fungsi adalah integer n, return True jika n ada pada range +-10 dari 100 atau 200. Selebihnya berarti akan menghasilkan return value False.

Contoh output:

      plus_minus(90) --> True
      plus_minus(150) --> False
      plus_minus(104) --> True

```python
# Insert your code here

def plus_minus(angka):
  return (abs(100 - angka) <= 10) or (abs(200 - angka) <= 10)

print(plus_minus(90))
print(plus_minus(150))
print(plus_minus(104))
```

Buatlah function yang dapat mengecek sebuah angka apakah angka tersebut berada dalam range angka yang diberikan atau tidak.

Contoh:

- Input:
      dalam_range(21, 10, 100)

- Output: 
      True

**NB:**

Parameter pertama adalah angka yang ingin dicari apakah dalam range atau tidak.

```python
# Insert your code here

def dalam_range(angka, bawah, atas):
  if angka in range(bawah, atas):
    print(f'{angka} ada dalam range antara {bawah} dan {atas}.')
  else:
    print(f'{angka} berada di luar range antara {bawah} dan {atas}.')

dalam_range(401, 200, 1000)
```
