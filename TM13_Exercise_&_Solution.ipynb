**1. Buatlah function untuk mengecek apakah sebuah string hanya mengandung alphanumeric atau bukan!**

```python
'ABCDEFabcdef123450'
'*&%@#!}{'
```

```python
# Insert your code here
def alphanumeric(string):
  string = re.search(r'[a-zA-Z0-9]', string) # alternatif lain menggunakan \w
  return bool(string)
```

```python
print(alphanumeric('ABCDEFabcdef123450'))
print(alphanumeric('*&%@#!}{'))
```

**2. Buatlah function yang dapat mencocokkan apakah sebuah string mengandung huruf 'a' dan diikuti oleh huruf 'b' sebanyak 0 atau lebih repetisi!**

```python
'ac'
'abc'
'abbc'
```

```python
# Insert your code here
def text_match(string):
  pattern = 'ab+'
  if re.search(pattern, string):
    return True
  else:
    return False
```

```python
print(text_match('ac'))
print(text_match('abc'))
print(text_match('abbc'))
print(text_match('saya belajar regex'))
```

**3. Buatlah function yang dapat mencocokkan apakah string memiliki huruf 'm' dan diikuti oleh 3 huruf 'n'!**

```python
'mnnn'
'mmnno'
```

```python
# Insert your code here
def m_followed(string):
  pattern = 'mn{3}'
  if re.search(pattern, string):
    return True
  else:
    return False
```

```python
print(m_followed('mnnn'))
print(m_followed('mmnno'))
```

**4. Buatlah function yang dapat mencocokkan apakah string memiliki huruf 'r' dan diikuti oleh 2 hingga 3 huruf 's'!**

```python
'rs'
'rrssssst'
```

```python
# Insert your code here
def r_followed(string):
  pattern = 'rs{2,3}?'
  if re.search(pattern, string):
    return True
  else:
    return False
```

```python
print(r_followed('segars'))
print(r_followed('segarss'))
print(r_followed('segarssss'))
```

**5. Buatlah function untuk dapat mencocokkan rangkaian kata dengan huruf kecil yang disambungkan dengan underscore!**

```python
'aab_cbbbc'
'aab_Abbbc'
'Aaab_abbbc'
```

```python
# Insert your code here
def lowercase_underscore(string):
  pattern = '^[a-z]+_[a-z]+$'
  if re.search(pattern, string):
    return True
  else:
    return False
```

```python
print(lowercase_underscore('aab_cbbbc'))
print(lowercase_underscore('aab_Abbbc'))
print(lowercase_underscore('Aaab_abbbc'))
```

**6. Buatlah function yang dapat mencocokkan sebuah string memiliki huruf 'a' diikuti oleh karakter apapun dan karakter terakhirnya adalah huruf 'b'!**

```python
'aabbbbd'
'aabAbbbc'
'accddbbjjjb'
```

```python
# Insert your code here
def a_b(string):
  pattern = 'a.+b$'
  if re.search(pattern, string):
    return True
  else:
    return False
```

```python
print(a_b('aabbbbd'))
print(a_b('aabAbbbc'))
print(a_b('accddbbjjjb'))
print(a_b('ab')) # False karena setelah a tidak diikuti oleh karakter lain. 
                 # Minimal harus ada 1 repetisi.
```

**7. Buatlah function yang dapat mencocokkan bahwa sebuah string dimulai dengan sebuah kata!**

```python
'Kemarin paman datang membawa banyak bingkisan.'
' Kemarin paman datang membawa banyak bingkisan.'
```

```python
# Insert your code here
def beginning(string):
  pattern = '^\w+'
  if re.search(pattern, string):
    return True
  else:
    return False
```

```python
print(beginning('Kemarin paman datang membawa banyak bingkisan.'))
print(beginning(' Kemarin paman datang membawa banyak bingkisan.'))
```

**8. Buatlah function yang dapat mencocokkan bahwa suatu string mengandung huruf 'x'!**

```python
'Latihan regex.'
'Latihan Python.'
```

```python
# Insert your code here
def find_x(string):
  pattern ='\w*x|X.*'
  if re.search(pattern, string):
    return True
  else:
    return False
```

```python
print(find_x('Latihan regex'))
print(find_x('Latihan Python'))
```

**10. Buatlah function yang dapat menghapus angka 0 pada IP address berikut!**

```python
ip = '123.01.000000044000000.136'
```

```python
# Hanya menghapus angka 0 setelah titik.
# Insert your code here
def remove_zero(string): 
  pattern = re.sub('\.[0]*', '.', string)
  return pattern
```

```python
remove_zero(ip)
```

```python
# Menghapus semua angka 0 yang ada pada target string.
def remove_all_zero(string): 
  pattern = re.sub(r'[0]', '', string)
  return pattern
```

```python
remove_all_zero(ip)
```

**11. Buatlah function untuk mencari sebuah kata spesifik dalam sebuah string. Selain itu, tampilkan juga lokasi (index) di mana pattern tersebut ditemukan!**

- Teks: 'Semua anak kucing itu berbulu putih hitam.'
- Kata yang dicari: 'kucing'

```python
# Insert your code here
pattern = 'kucing'
teks = 'Semua anak kucing itu berbulu putih hitam.'

def find_word(pattern, teks):
  match = re.search(pattern, teks)
  index_awal = match.start()
  index_akhir = match.end() - 1
  
  return print(f'''Kata '{match.re.pattern}' 
  ditemukan pada '{match.string}', pada 
  index {index_awal} hingga {index_akhir}''')
```

```python
find_word(pattern, teks)
```

**12. Buatlah function untuk mengonversi format date dari yyyy-mm-dd ke format dd-mm-yyyy!**

```python
# Insert your code here
def change_format(dt):
  # Memanfaatkan group reference 
  # (angka posisi dari suatu grup)
  return re.sub(r'(\d{4})-(\d{1,2})-(\d{1,2})', 
                  '\\3-\\2-\\1', dt) 
  
```

```python
date_today = '2022-6-13'
print('Format awal YYYY-MM-DD:', 
    date_today)
print('Format baru DD-MM-YYYY:', 
    change_format(date_today))
```

**13. Buatlah function yang dapat mengambil angka saja dari string berikut!**

string = 'Sepuluh 10, Lima puluh 50, Seratus 100'

```python
# Insert your code here
def take_numbers(string):
  result = re.split('\D+', string)
  for element in result:
    print(element)
```

```python
take_numbers('Sepuluh 10, Lima puluh 50, Seratus 100')
```

```python
def take_numbers_list(string):
  result = re.split('\D+', string)
  output = []
  for element in result:
    if element == '':
      continue
    else:
      output.append(element)
  # print(*output, sep='\n')
  print('\n'.join(output))
```

```python
take_numbers_list('Sepuluh 10, Lima puluh 50, Seratus 100')
```
