# String Indexing & Slicing

```python
sentence = 'Today, I am learning how to code in Python.'
```

```python
# Hanya mengambil karakter index ke-3 saja dari 
    # variabel sentence
print(sentence[3])
```

```python
# Titik dua setelah angka berarti kita ingin mengambil 
    # semua karakter dimulai dari index karakter yang kita 
    # definisikan.
# Angka index yang didefinisikan bersifat inclusive 
    # (termasuk ke dalam output).
print(sentence[3:])
```

```python
# Titik dua sebelum angka berarti kita ingin mengambil 
    # karakter hingga sebelum angka index yang didefinisikan.
# Angka index yang didefinisikan bersifat exclusive 
    # (tidak termasuk ke dalam output).
print(sentence[:3])
```

```python
# Start : Stop : Step
# Inclusive : Exclusive : Default is 1
# In Python, indexing starts from 0
print(sentence[2:10]) # Mengambil karakter dari index ke-2 hingga index ke-9
```

```python
# Negative indexing
# Indexing dari kanan ke kiri, dimulai dari -1
print(sentence[-2]) # n
print(sentence[-2:-5]) # Tidak error, tapi tidak 
                    # mengembalikan hasil apapun.
print(sentence[-5:-2]) # tho
print(sentence[:]) # return the whole string
print(sentence[::]) # return the whole string
print(sentence[::2]) # Menggunakan step 2, jadi ada 
                        # 1 karakter yang dilewat.
```

# Additional String Methods

```python
sentence = 'Today, I am learning how to code in Python.'
```

```python
# Count untuk menghitung jumlah karakter yang ingin 
    # dicari pada suatu string
print(sentence.count('a'))
```

```python
# Endswith untuk mengecek apakah suatu string diakhiri 
    # dengan suatu kata atau karakter yang kita definisikan.
print(sentence.endswith('Python.'))
```

```python
# isalnum: return True jika string hanya berisi alphabet 
    # dan numerik (whitespace tidak termasuk)
print(sentence.isalnum())
```

```python
# isalpha: return True jika string hanya berisi alphabet
print(sentence.isalpha())
```

```python
# islower: return True jika semua huruf pada string 
    # adalah lower case
print(sentence.islower())
```

```python
# isupper: return True jika semua huruf pada string 
    # adalah upper case
print(sentence.isupper())
```

```python
# isspace: return True jika string hanya memiliki 
    # whitespace(s)
print(sentence.isspace())
```

```python
# replace()
# Mengganti karakter tertentu dengan nilai baru
print(sentence.replace('a', 'A'))
```

```python
# Mengganti karakter hanya sesuai dengan jumlah 
    # yang diinginkan
print(sentence.replace('a', 'A', 2))
```
