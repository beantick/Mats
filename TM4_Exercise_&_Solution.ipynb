**1.**

 Buat algoritma untuk menentukan apakah suatu bilangan adalah bilangan prima.

 Contoh:

- Input:
        Masukkan angka: 7

- Output:
        7 adalah bilangan prima

- Input:
        Masukkan angka: 12
        
- Output:
        12 bukan merupakan bilangan prima

```python
# Insert your code here

angka = int(input('Masukkan angka: '))

# Bilangan prima dimulai dari > 1
# Bilangan prima hanya bisa dibagi dengan dirinya sendiri
if angka > 1:
  for i in range(2, angka): 
    if angka % i == 0: 
      print(f'{angka} bukan merupakan bilangan prima')
      break
  else:
    print(f'{angka} merupakan bilangan prima')

else:
  print(f'{angka} bukan merupakan bilangan prima')
    
```

Dari list berikut: [12, 15, 1, 7, 4, 100]

Buat algoritma untuk mencari angka tertinggi di dalam list tanpa menggunakan fungsi max atau sort pada Python.

- Output:
      100 

```python
# Insert your code here

list_angka = [12, 15, 1, 7, 4, 100]

angka_tertinggi = list_angka[0]

for angka in list_angka:
  if angka > angka_tertinggi:
    angka_tertinggi = angka

print(angka_tertinggi)
```

- Ketika angka 13 ketemu, print 'angka 13 ditemukan'

```python
# Insert your code here
x = 1
angka_dicari = 13

while x <= 100:
  print(x)

  if x == angka_dicari:
    print(f'\nAngka {angka_dicari} ditemukan')
    break
  x += 1
```

- Gunakan looping untuk membuat list yang berisikan huruf depan dari setiap kata pada kalimat di bawah ini:

**kalimat = ''Buatlah list yang berisikan huruf depan dari setiap kata pada kalimat ini'**


- Gabungkan setiap huruf depan dari semua kata menjadi satu kata!

- Output:
      Blybhddskpki

```python
# Insert your code here
kalimat = 'Buatlah list yang berisikan huruf depan dari setiap kata pada kalimat ini'

hasil = []

for kata in kalimat.split():
  # hasil.append(kata[0])
  hasil += kata[0]
print(''.join(hasil))
```

Lanjutkanlah program 'Fruits market' yang sudah dibuat pada latihan-latihan sebelumnya.

Untuk latihan kali ini, tambahkanlah beberapa fitur, yaitu:

**1. Tambahkan fitur 'stock'**

     - Setiap buah memiliki jumlah stock masing-masing.
     - Jumlah stock dari setiap buah bebas berjumlah berapa saja.

Saat user memasukkan jumlah buah yang ingin dibeli melebihi jumlah stock yang ada, maka akan muncul di output berupa informasi bahwa stock kurang dari permintaan, dan meminta user untuk memasukkan jumlah buah yang ingin dibelinya kembali, sesuai dengan jumlah stock yang ada.

Program juga tidak akan berpindah ke proses selanjutnya jika user tidak memasukkan jumlahnya dengan tepat.

**2. Update fitur 'pembayaran'**

Dengan menggunakan looping, jika user memasukkan jumlah yang kurang dari seharusnya, akan muncul teks di output berupa informasi bahwa uang yang dimasukkan masih kurang, dan meminta user untuk memasukkan ulang jumlah uangnya.

Untuk informasi teksnya:

    - Jika uang kurang dari total --> Munculkan 'jumlah kekurangan yang harus dibayar'.
    - Jika uang lebih dari total --> Tampilkan 'terima kasih dan jumlah kembaliannya'.
    - Jika uang = total --> Tampilkan 'terima kasih' saja.
    

Aplikasi akan berhenti atau selesai jika user memasukkan jumlah uang yang sama atau lebih dari total harga.

```python
# Insert your code here

# Harga buah
harga_apel = 10000
harga_jeruk = 15000
harga_anggur = 20000

# Stock buah
stock_apel = 5
stock_jeruk = 7
stock_anggur = 10

# Input jumlah buah yang ingin dibeli user
while True:
  jumlah_apel = int(input('Masukkan jumlah apel yang ingin dibeli: '))
  if (jumlah_apel > stock_apel):
    print(f'Jumlah yang dimasukkan terlalu banyak. \n\
      Stock apel yang tersedia hanya {stock_apel}.')
  else:
    break

while True:
  jumlah_jeruk = int(input('Masukkan jumlah jeruk yang ingin dibeli: '))
  if (jumlah_jeruk > stock_jeruk):
    print(f'Jumlah yang dimasukkan terlalu banyak. \n\
      Stock jeruk yang tersedia hanya {stock_jeruk}.')
  else:
    break

while True:
  jumlah_anggur = int(input('Masukkan jumlah anggur yang ingin dibeli: '))
  if (jumlah_anggur > stock_anggur):
    print(f'Jumlah yang dimasukkan terlalu banyak. \n\
      Stock anggur yang tersedia hanya {stock_anggur}.')
  else:
    break

# Kalkulasi total harga
total_harga_apel = jumlah_apel * harga_apel
total_harga_jeruk = jumlah_jeruk * harga_jeruk
total_harga_anggur = jumlah_anggur * harga_anggur
total_harga = total_harga_apel + total_harga_jeruk + total_harga_anggur

print(f'''
      Detail Belanja
      
      Apel: {jumlah_apel} x {harga_apel} = {total_harga_apel}
      Jerukl: {jumlah_jeruk} x {harga_jeruk} = {total_harga_jeruk}
      Anggur: {jumlah_anggur} x {harga_anggur} = {total_harga_anggur}

      Total: {total_harga}
      ''')

while True:
  jumlah_uang = int(input('Masukkan jumlah uang: '))

  if jumlah_uang > total_harga:
    kembalian = jumlah_uang - total_harga
    print(f'\nTerima kasih. \n\nUang kembalian Anda: {kembalian}')
    break

  elif jumlah_uang == total_harga:
    print('\nTerima kasih.')
    break
  else:
    kekurangan = total_harga - jumlah_uang
    print(f'Uang Anda kurang sebesar {kekurangan}. \n\n\
      Silakan masukkan jumlah uang yang sesuai.')
```

```python
# Insert your code here

baris = 5

for i in range(baris + 1, 1, -1): # 6 5 4 3 2
  for j in range(0, i - 1): # 0 1 2 3 4
    print('*', end=' ')
  print()

for i in range(1, baris): # 1 2 3 4
  for j in range(0, i + 1): # iterasi pertama 0, 2 --> 0 & 1
    print('*', end=' ')
  print()
```
