# Collection Data Types Exercise

**1.**

Temukan angka minimum dan maksimum pada list berikut tanpa menggunakan method min, max, dan sort:

- Input:
      angka = [31, 5, 1, 7, 88, 42]

- Output:
      Angka minimum = 1
      Angka maksimum = 88

```python
# Insert your code here

list_angka = [31, 5, 1, 7, 88, 42]

angka_minimum = list_angka[0]
angka_maksimum = list_angka[0]

for angka in list_angka:
  if (angka < angka_minimum):
    angka_minimum = angka
  elif (angka > angka_maksimum):
    angka_maksimum = angka

print(f'Angka minimum = {angka_minimum}')
print(f'Angka maksimum = {angka_maksimum}')
```

**2.**

Urutkan angka pada list berikut dari yang terkecil hingga terbesar tanpa menggunakan method sort!

- Input:
      list_angka = [31, 5, 1, 7, 88, 42, 202, 29]

- Output:
      list_angka_sorted = [1, 5, 7, 29, 31, 42, 88, 202]

```python
# Insert your code here

list_angka = [31, 5, 1, 7, 88, 42, 202, 29]

print(f'List angka sebelum diurutkan: {list_angka}')

for i in range(len(list_angka)):
  for j in range(i + 1, len(list_angka)):
    if (list_angka[i] > list_angka[j]):
      list_angka[i], list_angka[j] = list_angka[j], list_angka[i]

print(f'List angka setelah diurutkan: {list_angka}')
```

**3.**

Gunakan user input, buat program untuk mencari berapa kali tiap kata muncul dalam 1 kalimat.

Contoh:

- Input:
      kalimat = 'Aku baru makan nasi terus aku mau makan mie nanti malam'

- Output: 
      Jumlah kata 'Aku' ada sebanyak 2
      Jumlah kata 'Baru' ada sebanyak 1
      Jumlah kata 'Makan' ada sebanyak 2
      Jumlah kata 'Terus' ada sebanyak 1
      Jumlah kata 'Mau' ada sebanyak 1
      Jumlah kata 'Mie' ada sebanyak 1
      Jumlah kata 'Nanti' ada sebanyak 1
      Jumlah kata 'Malam' ada sebanyak 1


```python
# Insert your code here

kalimat = str(input('Masukkan kalimat: '))

kata = {} # key-value pair

for word in kalimat.lower().split():
  if word in kata.keys():
    kata[word] += 1
  else:
    kata[word] = 1

for key, val in kata.items():
  print(f"Jumlah kata '{key.capitalize()}' ada sebanyak {val}")
```

Input:

Masukkan 5 film, dipisah dengan koma untuk film kesukaan kamu dan kesukaan teman kamu (gunakan user input).

Output: 
Presentase kesamaan film yang kalian suka.



**Contoh**

- Input:
        'Masukkan 5 film kesukaan Anda dipisahkan dengan tanda koma: '

        'Masukkan 5 film kesukaan teman Anda dipisahkan dengan tanda koma: '

- Output:
        'Kesukaan film kalian yang sama sebesar xx %'

```python
# Insert your code here

is_valid = True

while is_valid:
  user1 = str(input('Masukkan 5 film kesukaan kamu, tiap film dipisahkan dengan tanda koma: '))

  user1_list = user1.split(', ')
  user1_len = len(user1_list)

  user2 = str(input('Masukkan 5 film kesukaan teman kamu, tiap film dipisahkan dengan tanda koma: '))

  user2_list = user2.split(', ')
  user2_len = len(user2_list)

  if user1_len and user2_len == 5:
    print(f'\n5 film kesukaan Anda adalah {user1}')
    print(f'5 film kesukaan teman Anda adalah {user2}')

    is_valid = False

  elif user1_len and user2_len > 5:
    print('Anda memasukkan judul film lebih dari 5. Silakan masukkan hanya 5 judul film.')
  elif user1_len and user2_len < 5:
    print('Anda memasukkan judul film kurang dari 5. Silakan masukkan 5 judul film.')

set1 = set(user1_list)
set2 = set(user2_list)

kesamaan = set1.intersection(set2)
jumlah = len(kesamaan)

if jumlah == 5:
  print('\nKesukaan film kalian yang sama sebesar 100%')
elif jumlah == 4:
  print('\nKesukaan film kalian yang sama sebesar 80%')
elif jumlah == 3:
  print('\nKesukaan film kalian yang sama sebesar 60%')
elif jumlah == 2:
  print('\nKesukaan film kalian yang sama sebesar 40%')
elif jumlah == 1:
  print('\nKesukaan film kalian yang sama sebesar 20%')
else:
  print('Kalian tidak memiliki kesamaan dalam film yang disukai')
```

- Output: 

      [2, 3, 4, 5, 6, 7, 8, 9, 10, 12, 14, 15, 16, 18, 20, 21, 22, 24, 25, 26,
      27, 28, 30, 32, 33, 34, 35, 36, 38, 39, 40, 42, 44, 45, 46, 48, 49, 50,
      51, 52, 54, 55, 56, 57, 58, 60, 62, 63, 64, 65, 66, 68, 69, 70, 72, 74,
      75, 76, 77, 78, 80, 81, 82, 84, 85, 86, 87, 88, 90, 91, 92, 93, 94,
      95, 96, 98, 99, 100]

```python
# Insert your code here

# Menggunakan list comprehension
pembagian = [angka for angka in range(1, 101) if True\
     in [True for pembagi in range(2, 10)\
        if angka % pembagi == 0]]
print(pembagian)
print(len(pembagian))
```

```python
# Menggunakan for loop

pembagian = []

for angka in range(1, 101):
  for pembagi in range(2, 10):
    if angka % pembagi == 0:
      pembagian.append(angka)

pembagian = set(pembagian)
print(pembagian)
print(len(pembagian))
```

Dengan menggunakan list comprehension, buatlah kode yang hanya mengambil angka pada suatu kalimat.

**Contoh kalimat:** 'Kemarin dari jam 2 hingga 3.30 sore saya bertemu dengan 10 orang di 4 tempat berbeda'

- Output: 
      ['2', '3.30', '10', '4']

```python
# Insert your code here

kalimat = 'Kemarin dari jam 2 hingga 3.30 sore saya bertemu dengan 10 orang di 4 tempat berbeda'

kata = kalimat.split()

hasil = [angka for angka in kata if not angka.isalpha()]
print(hasil)
```

# Fruits Market App

Dari app furits market yang sudah dikerjakan, tambahkan fitur-fitur sebagai berikut:

- Yang pertama, sekarang user tidak hanya bisa belanja, namun ada 3 pilihan tambahan, yaitu menampilkan daftar buah, menambah buah, dan menghapus buah.
- Jadi, programnya akan memiliki 5 menu utama:
    1. Menampilkan daftar buah
    2. Menambah buah
    3. Menghapus buah
    4. Membeli buah
    5. Exit Program

- Yang kedua, saat pertama kali program dimulai, daftar buah sudah berisikan 3 buah beserta stock dan harganya.

- Yang ketiga, untuk setiap selesai menambah dan menghapus buah, program akan menampilkan daftar buah terbaru. 

- Yang keempat, saat proses membeli, gunakanlah sistem seperti keranjang belanja atau cart.

- Yang kelima, seperti program market sebelumnya:

    - Kalau uangnya lebih kecil dari total harga belanja, user akan diminta input ulang sampai uangnya cukup.

- Yang keenam, saat selesai membeli buah, jangan lupa untuk mengurangi stock dari buah-buah yang dibeli, dan kemudian kosongkan cart-nya
- Yang ketujuh, setiap selesai menggunakan suatu menu, program akan kembali ke main menu, dan program hanya akan berhenti bila user menjalankan menu Exit Program.


Hints:
  - Gunakanlah Python List untuk menyimpan data dari buah-buah, supaya nanti dapat menambahkan buah baru dan menghapus buah yang ada. 

Otomatis di program ini juga membutuhkan Python Conditional Statement dan Looping Statement.

```python
# Insert your code here

list_buah = [
    ['Apel', 20, 10000],
    ['Jeruk', 15, 15000],
    ['Anggur', 25, 20000]
]

cart = []

while True:
    pilihan_menu = input('''
        Selamat Datang di Pasar Buah

        List Menu:
        1. Menampilkan Daftar Buah
        2. Menambah Buah
        3. Menghapus Buah
        4. Membeli Buah
        5. Exit Program

        Masukkan angka menu yang ingin dijalankan: ''')

    if(pilihan_menu == '1'):

        print('Daftar Buah\n')
        print('Index\t| Nama  \t| Stock\t| Harga')
        
        for i in range(len(list_buah)):
            print(f'{i}\t| {list_buah[i][0]}  \t| {list_buah[i][1]}\t| {list_buah[i][2]}')

    elif(pilihan_menu == '2'):

        nama_buah = input('Masukkan nama buah: ')
        stock_buah = int(input('Masukkan stock buah: '))
        harga_buah = int(input('Masukkan harga buah: '))
        
        list_buah.append([nama_buah, stock_buah, harga_buah])
        
        print('Daftar Buah\n')
        print('Index\t| Nama  \t| Stock\t| Harga')
        
        for i in range(len(list_buah)):
            print(f'{i}\t| {list_buah[i][0]}  \t| {list_buah[i][1]}\t| {list_buah[i][2]}')

    elif(pilihan_menu == '3'):

        print('Daftar Buah\n')
        print('Index\t| Nama  \t| Stock\t| Harga')
        
        for i in range(len(list_buah)):
            print(f'{i}\t| {list_buah[i][0]}  \t| {list_buah[i][1]}\t| {list_buah[i][2]}')

        index_buah = int(input('Masukkan index buah yang ingin dihapus: '))
        
        del list_buah[index_buah]

        print('Daftar Buah\n')
        print('Index\t| Nama  \t| Stock\t| Harga')
        
        for i in range(len(list_buah)):
            print(f'{i}\t| {list_buah[i][0]}  \t| {list_buah[i][1]}\t| {list_buah[i][2]}')


    elif(pilihan_menu == '4'):

        print('Daftar Buah\n')
        print('Index\t| Nama  \t| Stock\t| Harga')

        for i in range(len(list_buah)):
            print(f'{i}\t| {list_buah[i][0]}  \t| {list_buah[i][1]}\t| {list_buah[i][2]}')
        
        while True:
            index_buah = int(input('Masukkan index buah yang ingin dibeli: '))
            
            if index_buah not in range(0, len(list_buah)):
                print('Index buah tidak ditemukan, silakan masukkan angka index yang sesuai.')
                continue
            else:
                pass
            
            jumlah_buah = int(input('Masukkan jumlah yang ingin dibeli: '))
            
            if(jumlah_buah > list_buah[index_buah][1]):
                print(f'Stock tidak cukup, sisa stock {list_buah[index_buah][0]} adalah {list_buah[index_buah][1]}')
            else:
                cart.append([list_buah[index_buah][0], jumlah_buah, list_buah[index_buah][2], index_buah])
            
            for item in cart:
                list_buah[item[3]][1] -= jumlah_buah
            
            print('Isi Cart:')
            print('Nama\t| Jumlah\t| Harga')
            
            for item in cart:
                print(f'{item[0]}\t| {item[1]}\t\t| {item[2]}')

            break

        while True:
            checker = str(input('Mau beli yang lain? (Ya/Tidak) = '))

            if (checker.lower() == 'ya'):
                index_buah = int(input('Masukkan index buah yang ingin dibeli: '))

                if index_buah not in range(0, len(list_buah)):
                    print('Index buah tidak ditemukan, silakan masukkan angka index yang sesuai.')
                    continue
                else:
                    pass
                
                print(f'Sisa stock {list_buah[index_buah][0]} adalah {list_buah[index_buah][1]}.')

                if(list_buah[index_buah][1] == 0):
                    print('Maaf, stock buah yang dipilih sedang kosong. Silakan pilih buah yang lain jika mau.')
                    continue
                else:
                    jumlah_buah_tambahan = int(input('Masukkan jumlah yang ingin dibeli: '))
                    

                if(jumlah_buah_tambahan > list_buah[index_buah][1]):
                    print(f'Stock tidak cukup, masukkan jumlah yang sesuai dengan sisa stock.')
                else:
                    cart.append([list_buah[index_buah][0], jumlah_buah_tambahan, list_buah[index_buah][2], index_buah]) 
                    list_buah[index_buah][1] -= jumlah_buah_tambahan
            
            elif (checker.lower() != 'ya' and checker.lower() != 'tidak'):
                continue
            else:
                break

            
        print('Daftar Belanja:')
        print('Nama\t| Jumlah\t| Harga\t| Total Harga')
        total_harga = 0

        for item in cart:
            print(f'{item[0]}\t| {item[1]}\t| {item[2]}\t| {item[1] * item[2]}')
            total_harga += item[1] * item[2]    
        
        while True:
            print(f'Total yang harus dibayar = {total_harga}')
            if total_harga != 0:
                jumlah_uang = int(input('Masukkan jumlah uang: '))

                if(jumlah_uang > total_harga):
                    kembali = jumlah_uang - total_harga
                    print(f'Terima kasih \n\nUang kembalian Anda: {kembali}')
                    cart.clear()
                    break
            
                elif(jumlah_uang == total_harga):
                    print('Terima kasih')
                    cart.clear()
                    break
            
                else:
                    kekurangan = total_harga - jumlah_uang
                    print(f'Uang Anda kurang sebesar {kekurangan}')

            else:
                print(f'Terima kasih telah berkunjung.')
                break
            
            
                
    elif(pilihan_menu == '5'):
        break
```
