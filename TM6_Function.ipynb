# Function

- Jika suatu proses dilakukan berulang kali, maka kita bisa menggunakan Function.
- Function bisa digunakan di banyak tempat berbeda, agar tidak perlu copy paste.
- Function bisa digunakan oleh orang lain tanpa harus orang tersebut tahu proses apa yang terjadi di dalam Function tersebut.

- Built-in function: function bawaan dari Python.

- Function adalah sebuah blok kode yang terorganisir, memiliki nama, dapat menerima input dan mengeluarkan output, dan dapat dijalankan berulang kali. 


# Regular Function

```python
# Regular function

def tambah(angka1, angka2):
  return angka1 + angka2

tambah(3, 5)

# Lambda function (funtion tanpa nama)
# Biasanya hanya digunakan satu kali pakai saja

lambda angka1, angka2: angka1 + angka2 
```

## Function Without Input & Output

```python
def salam():
  print('Halo, selamat datang!')
  print('Silakan bersenang-senang.')

print(salam())

# Untuk mengecek bahwa suatu regular function tidak memiliki output, bisa tambahkan print() dalam memanggil nama functionnya.
# Hasilnya akan berupa 'None'.
```

## Function With Input but Without Output

```python
def perkenalan(nama, umur):
  print(f'Halo semua, perkenalkan nama saya {nama}, umur saya {umur} tahun.')
  print('Senang berkenalan dengan Anda.')

# function(parameter=argument, parameter=argument)
perkenalan('Aulia', 17) # Arguments bisa diinput tanpa menuliskan nama tiap parameternya, tapi urutannya harus sesuai agar outputnya juga sesuai.

# Jika urutannya ingin acak, kita bisa menuliskan nama parameternya saja.
perkenalan(umur=17, nama='Andi')
```

## Default Parameter

- Parameter tanpa default value tidak bisa didefinisikan setelah parameter yang memiliki default value.
- Contoh: angka3 tidak bisa didefinisikan setelah angka2 karena angka2 memiliki default value.


```python
def tambah(angka1, angka3, angka2=10):
  return angka1 + angka2 + angka3

tambah(1, 10.5, angka2=2) # 1 sebagai angka1, 10.5 sebagai angka3. Urutan harus sesuai.
                          # Dapat menerima int atau float
```

```python
def perkenalan(nama='Unknown', umur=0):
  if (nama == 'Unknown' and umur > 0):
    print(f'Halo, saya berusia {umur} tahun.')
  elif (nama != 'Unknown' and umur == 0):
    print(f'Halo, nama saya {nama}.')
  elif (nama != 'Unknown' and umur > 0):
    print(f'Halo, nama saya {nama} dan saya berumur {umur} tahun.')
  else:
    print('Halo, saya tidak mau memberitahukan nama dan umur saya.')

  print('Senang bertemu dengan teman-teman.')

perkenalan()
```

```python
perkenalan(umur=30)
```

```python
perkenalan(nama='Aaron')
```

```python
perkenalan(umur=33, nama='Alice')
```

# Local & Global Variable

```python
def total(x, y):
  z = x + y
  return z

total(4, 5)
```

```python
# Error, karena z adalah local variable.
print(z)
```

## Local Variable


- Variabel yang dideklarasikan dalam suatu blok pemrograman. 
- Local Variable hanya dapat digunakan di dalam blok kode di mana dia dideklarasikan/dibuat. 
- Local Variable akan ada selama blok kode/function masih dijalankan. Setelah itu, Local Variable akan dihapus secara otomatis. 

```python
def coba():
  print(x + 2)
  return x + 3

x = 5
print(coba(), x)
```

## Global Variable

- Variabel yang didefinisikan di luar function. 
- Global Variable memiliki cakupan global yang berarti memegang nilainya sepanjang masa program dijalankan. 
- Global Variable dapat diakses di seluruh program oleh function apa pun yang terdapat di dalam program.


```python
def coba():
  x = 100 # local variable
  print(x + 2)
  return x + 3

x = 5 # global variable

print(coba(), x)
print()
print(coba())
```

```python
x = 5

def coba():
  x += 7
  print(x + 2)
  return x + 3

print(coba(), x) # Error, karena x di dalam function tidak bisa mereferensikan ke x di luar function. 

# Agar tidak error, salah satu caranya adalah dengan menggunakan keyword 'global' di dalam function.
```

```python
# Cara 1
# Membuat local variable x = 2

def coba():
  x = 2 # local variable
  x += 7
  print(x + 2)
  return x + 3

x = 5 # global variable
print(coba(), x)
```

```python
# Cara 2
# Define global x

def coba():
  global x # x akan bernilai 5
  x += 7
  print(x + 7)
  return x * 4

x = 5 # global variable

print(coba(), x) # nilai x menjadi 12 secara permanen karena sudah dikenakan operasi di dalam function yang menggunakan keyword 'global'.
```
