# Callback Function 

```python
# Function yang menerima argumen berupa function lain.

def add(num1, num2):
  return num1 + num2

def multiply(num1, num2):
  return num1 * num2

def calculate(fn, num1, num2):
  result = fn(num1, num2)
  return result

# Argumen pertama berupa function
print(calculate(add, 10, 5))
print(calculate(multiply, 10, 5))
```

```python
# Membuat program yang menghitung 
  # keliling persegi dan lingkaran
# Keliling persegi = 4 * s
# Keliling lingkaran = pi * 2r

import math

def keliling_persegi(s):
  return 4 * s

def keliling_lingkaran(r):
  return math.pi * 2 * r

def hitung_keliling(fn, ukuran):
  return fn(ukuran)

print(hitung_keliling(keliling_persegi, 4))
print(hitung_keliling(keliling_lingkaran, 4))
```

# Calling Other Function(s)

```python
# Function yang digunakan di dalam function lainnya.

def show_answer(answer):
  print(f'The answer is {answer}')

def multiply(num1, num2):
  result = num1 * num2
  show_answer(result)

def division(num1, num2):
  result = num1 / num2
  show_answer(result)

multiply(10, 5)
division(10, 5)
```

```python
# Membuat program yang menghitung 
  # keliling persegi dan lingkaran
# Keliling persegi = 4 * s
# Keliling lingkaran = pi * 2r

# Menggunakan callback function dan calling 
  # other function secara bersamaan.

import math

def show_answer(answer):
  print(f'Kelilingnya adalah {answer}')

def keliling_persegi(s):
  return 4 * s

def keliling_lingkaran(r):
  return math.pi * 2 * r

def hitung_keliling(fn, ukuran):
  return show_answer(fn(ukuran))

hitung_keliling(keliling_persegi, 4)
hitung_keliling(keliling_lingkaran, 4)
```

# Recursive Function

```python
# Sebuah function memanggil dirinya 
  # sendiri di dalam function tersebut.

# Recursive function

import time 

def timer(t):
  if t == 0:
    print('Time is over')
  else:
    print(t)
    time.sleep(1) # Delay untuk memunculkan hasil 
                  # iterasi selanjutnya sesuai dengan 
                  # lama waktu yang diinginkan
    return (timer(t - 1)) # recursive part

timer(5)
```

```python
# Menggunakan while loop

def timer(t):
  while t > 0:
    print(t)
    t -= 1
  print('Time is over')

timer(5)
```

# Lambda Function

- Function yang tidak memiliki nama.
- Dapat memiliki sejumlah parameter, tapi hanya dapat memiliki 1 expression.
- Kalau condition yang dibutuhkan cukup kompleks, lebih baik gunakan regular function.
- Meskipun lambda function adalah function tanpa nama, namun kita tetap bisa menyimpannya ke dalam sebuah variable.

- Syntax:

      lambda parameters: expression

```python
# Contoh 1

kuadrat = lambda angka: angka**2
print(kuadrat(10))
```

```python
# Contoh 2
tambah = lambda num1, num2, num3: num1 + num2 + num3

print(tambah(10, 20, 30))
```

# Map Function

```python
# Syntax

# map(function, iterable, ...)
```

```python
# Digunakan untuk mengubah nilai pada 
  # suatu collection data type.

# Contoh 1
list_angka = [2, 3, 4, 5]

# Dengan regular function
def kuadrat(angka):
  return angka**2

hasil_map = map(kuadrat, list_angka)
print(hasil_map) # Return map class object

# Tambahkan keyword collection data type 
  # untuk menampilkan hasil map function.
print(list(hasil_map))

# Dengan lambda function
hasil_map_lambda = list(map(lambda angka: angka**2, list_angka))
print(list(hasil_map_lambda))
```

```python
# Contoh 2

# Function untuk membuat duplikat

def map_duplikat(fn, collection):
  new_collection = []

  for item in collection:
    new_collection.append(fn(item))

  return new_collection

list_angka = [1, 2, 3, 4, 5]

# Buat function untuk menambahkan kata 
#   'Angka' sebelum tiap angka di list_angka

def ubah(angka):
  return f'Angka {angka}'

# Gunakan function map_duplikat() dengan 
  # argumen berupa function ubah dan list_angka
new_list = map_duplikat(ubah, list_angka)
print(list(new_list))
```

```python
# Contoh 2 dengan map function
hasil_map = map(ubah, list_angka)
print(list(hasil_map))
```

```python
# Contoh 2 dengan lambda function
hasil_lambda = map_duplikat(lambda angka: f'Angka {angka}', list_angka)
print(hasil_lambda)
```

```python
# Contoh lain
num1 = [4, 5, 6]
num2 = [5, 6, 7]

result = map(lambda n1, n2: n1+n2, num1, num2)
print(list(result))
```

# Filter Function

```python
# Syntax

# filter(function, iterable)
```

```python
# Digunakan untuk filtering data yang mau ditampilkan.

salaries = [9300, 9500, 9200, 9800, 10300]

# Ingin menampilkan hanya salary yang 
  # lebih dari atau sama dengan 9500

# Cara for loop
salaries_filtered = []

for salary in salaries:
  if salary >= 9500:
    salaries_filtered.append(salary)

print(salaries_filtered)
```

```python
# Menggunakan filter function

# Regular function
def above_9500(salary):
  return salary >= 9500

print(list(filter(above_9500, salaries)))

# Lambda function
print(list(filter(lambda salary: salary >= 9500, salaries)))
```

```python
# Contoh 2

# Membuat function untuk melakukan filtering secara manual

def filter_manual(fn, collection):
  new_collection = []

  for item in collection:
    if fn(item):
      new_collection.append(item)

  return new_collection

def genap(angka):
  return angka % 2 == 0

list_angka = [1, 2, 3, 4, 5]

# Argument diisi dengan function genap
new_list = filter_manual(genap, list_angka)
print(list(new_list))
```

```python
# Menggunakan filter untuk contoh 2
hasil_filter = filter(genap, list_angka)
print(list(hasil_filter))
```

```python
# Menggunakan lambda function untuk contoh 2
hasil_lambda = filter_manual(lambda angka: angka % 2 == 0, list_angka)
print(list(hasil_lambda))
```

```python
# Contoh lain
list_huruf = ['a', 'x', 'k', 'u', 'i', 'm', 'o', 'n']

# a function that returns True if letter is vowel
def filter_vokal(huruf):
    vokal = ['a', 'e', 'i', 'o', 'u']
    return True if huruf in vokal else False

vokal_filtered = filter(filter_vokal, list_huruf)
print(vokal_filtered)

# converting to list
vokal_filtered = list(vokal_filtered)
print(vokal_filtered)
```

# *args & **kwargs

```python
## *args dan **kwargs

def myfunc(a, b):
    '''Function ini mengembalikan nilai 5% 
    dari jumlah perkalian kedua parameter'''
    return sum((a, b)) * 0.05

print(myfunc(40, 60)) # 5.0
```

```python
# Contoh 2

def myfunc(a, b, c=0, d=0, e=0):
    '''Function ini mengembalikan nilai 5% 
    dari jumlah perkalian kedua parameter'''
    return sum((a, b, c, d, e)) * 0.05

print(myfunc(40, 60, 100)) # 10.0
```

```python
print(myfunc(40, 60, 100, 100, 200, 300)) 
# Error karena jumlah argumen yang 
    # dimasukkan melebihi jumlah parameter
```

```python
# Di sinilah *args dapat berguna

def myfunc(*args):
    return sum((args)) * 0.05

print(myfunc(40, 60, 100, 100, 200, 300, 1000, 20000)) 
# Bisa memasukkan jumlah argumen tak terbatas
```

```python
# **kwargs (keyword args)

def myfunc(**kwargs):
    if 'buah' in kwargs:
        print(f"Pilihan buah saya adalah {kwargs['buah']}")
    else:
        print('Saya tidak menemukan buah pilihan saya.')

myfunc(buah='apel', sayur='kangkung')
```

```python
# Nama *args dan **kwargs bisa diubah seperti 
    # nama variabel sementara pada looping, 
    # tapi lebih baik
    # gunakan default (*args atau **kwargs) saja.
```

```python
def myfunc(*args, **kwargs):
    print(f"Saya mau {args[1]} {kwargs['cemilan']}")

myfunc(10, 20, 30, cemilan = 'risoles', dessert = 'puding')
```

```python
## Urutan penggunaan jika menggunakan *args dan **kwargs 
    # secara bersamaan, maka *args berada sebelum **kwargs
## Penggunaan **kwargs harus mendefinisikan nama parameternya.
```

```python
def myfunc(**kwargs, *args):
    print(f"Saya mau {args[0]} {kwargs['cemilan']}")

myfunc(cemilan = 'risoles', dessert = 'puding', 10, 20, 30) 
# Tidak bisa dijalankan, karena 
# invalid syntax (*args berada setelah **kwargs)
```

```python
def myfunc(*args, **kwargs):
    print(f"Saya mau {args[0]} {kwargs['cemilan']}")
    print(f"Saya mau {args[0]} {kwargs['cemilan'][0]}")

myfunc(10, 20, 30, cemilan = ['risoles', 'bolu'], dessert = 'puding')
```

```python
myfunc(10, 20, 30, cemilan = 'risoles', cemilan = 'bolu', dessert = 'puding') 
# Error, argumen yang sama tidak bisa diulang
```

```python
myfunc('cemilan', 20, 30, 'risoles', cemilan='puding') # Saya mau cemilan puding 
```
