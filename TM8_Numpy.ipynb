# Numpy

**Array:** 
- Tipe data terstruktur yang menyimpan data dengan tipe data yang seragam
- Mutable
- Index dimulai dari 0
- Bisa berbentuk 1D, 2D, 3D, ..., nD

**Dimensi:**
- Jumlah index yang dibutuhkan untuk memanggil suatu elemen
- array 1D punya 1 index --> vector
- array 2D punya 2 index --> matrix
- array 3D punya 2 index --> 3D array
- array nD punya 2 index --> nD array

**Numpy:**
- NumPy --> Numerical Python
- Library python untuk mengelola data array.
- Bisa digunakan untuk operasi matrix, aljabar linear, dan operasi matematika tingkat lanjut.
- Kecepatan operasi dengan array 50 kali lebih cepat daripada list.
- Function untuk NumPy Array lebih banyak untuk pekerjaan data science.

**Persamaan list dengan np.array**
1. Menggunakan []
2. Mutable
3. Can be indexed & sliced

**Perbedaan**
1. Array consumes less memory
2. Array can handle directly math ops
3. Array is faster
4. List can consist of different nested data size
5. List can store different dtypes

## Slicing

Hati-hati dengan slicing pada array, jika yang di-slice diubah nlainya, maka nilai pada array awal akan ikut berubah.

Solusinya adalah dengan menduplikat array terlebih dahulu dengan menggunakan .copy()

```python
arr = np.arange(0, 11)
arr
```

```python
# Slicing array dan simpan pada variabel baru
slice_arr = arr[0:5]
slice_arr
```

```python
# Mengganti nilai pada seluruh element
slice_arr[:] = 99
slice_arr
```

```python
# Kalau kita panggil lagi arr awal, nilai 
    # pada index 0 sampai 5 akan berubah juga
arr
```

```python
# Agar tidak mengubah nilai pada array awal, 
    # gunakan method np.copy()
arr_copy = arr.copy()
arr_copy
```

```python
arr_copy = arr_copy[0:5]
arr_copy
```

```python
arr_copy[:] = 100
arr_copy
```

```python
# Nilai pada array awal tidak berubah 
arr
```

```python
# Membuat array 2d

arr_2d = np.array([[10, 15, 20], 
[25,30, 35], [40, 45, 50]])
arr_2d
```

```python
# Slicing baris index 1
arr_2d[1]
```

```python
# Slicing baris index 0, lalu 
    # slicing kolom index 1
arr_2d[0][1] # [baris][kolom]
```

```python
# Slicing baris index 0, lalu 
    # slicing kolom index 1
arr_2d[0, 1] # [baris, kolom]
```

```python
# Slicing baris index 0 dan 1, 
    # lalu slicing kolom 1 dan 2
arr_2d[:2, 1:]
```

```python
# Hati-hati dalam melakukan indexing dengan slicing
# Kalau seperti ini, jadinya slicing baris index [0:2] 
    # terlebih dahulu, lalu slicing index [1:]
arr_2d[:2][:1]
```
